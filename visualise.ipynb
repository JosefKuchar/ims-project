```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('diameter.csv')

df['pcu'] = df['pcu'] / 10 # Convert to vehicles per hour

# Plot the data
sns.set_theme()
plt.figure(figsize=(5, 5))
sns.lineplot(x='diameter', y='pcu', data=df)
plt.ylim(0, 2500)
plt.xlim(20, 60)
plt.xlabel('Průměr okružní křižovatky (m)')
plt.ylabel('Kapacita (PCU/hod)')
plt.tight_layout()
plt.savefig('img/diameter.png', dpi=300)
plt.show()
```

```python

# Read in the data
df = pd.read_csv('approach-count.csv')

df['per_lane'] = (df['pcu'] / df['approach_count']) / 10

# Plot the data
sns.set_theme()
plt.figure(figsize=(5, 5))
sns.lineplot(x='approach_count', y='per_lane', data=df, marker='o')
plt.ylim(0, 1000)
plt.xlabel('Počet příjezdových jízdních pruhů')
plt.ylabel('Kapacita na jeden příjezdový pruh (PCU/hod)')
# Set xticks to integer values
plt.xticks(range(3, 7))
plt.tight_layout()
plt.savefig('img/approach-count.png', dpi=300)
plt.show()
```
