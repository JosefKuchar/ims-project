```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('diameter.csv')

df['pcu'] = df['pcu'] / 10 # Convert to vehicles per hour

# Plot the data
sns.set_theme()
plt.figure(figsize=(5, 5))
sns.lineplot(x='diameter', y='pcu', data=df)
plt.ylim(0, 2500)
plt.xlim(20, 45)
plt.xlabel('Průměr okružní křižovatky (m)')
plt.ylabel('Kapacita (pcu/hod)')
plt.show()
```
